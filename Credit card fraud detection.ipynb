## Using Under Sampling technique

```python
normal = data[data['Class']==0]
fraud = data[data['Class']==1]
```

```python
normal.shape
```

```python
fraud.shape
```

```python
normal_sample=normal.sample(n=473)
```

```python
new_data = pd.concat([normal_sample,fraud],ignore_index=True)
```

```python
new_data['Class'].value_counts()
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
X = new_data.drop('Class',axis=1)
y = new_data['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20, stratify=y, random_state=2)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_pred1 = log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred1)
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
precision_score(y_test,y_pred1)
```

```python
recall_score(y_test,y_pred1)
```

```python
f1_score(y_test,y_pred1)
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
y_pred2 = dt.predict(X_test)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
precision_score(y_test,y_pred2)
```

```python
recall_score(y_test,y_pred2)
```

```python
f1_score(y_test,y_pred2)
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
y_pred3 = rf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred3)
```

```python
precision_score(y_test,y_pred3)
```

```python
recall_score(y_test,y_pred3)
```

```python
f1_score(y_test,y_pred3)
```

```python
final_data = pd.DataFrame({'Models':['LR','DT','RF'],
                            "ACC":[accuracy_score(y_test,y_pred1),accuracy_score(y_test,y_pred2),accuracy_score(y_test,y_pred3)],
                            "Prec":[precision_score(y_test,y_pred1),precision_score(y_test,y_pred2),precision_score(y_test,y_pred3)],
                            "F_1":[f1_score(y_test,y_pred1),f1_score(y_test,y_pred2),f1_score(y_test,y_pred3)]})
```

```python
final_data
```

#### Logistic regressin is performing best on undersampling data

## Over sampling Using SMOTE

```python
X = data.drop('Class',axis=1)
y = data['Class']
```

```python
X.shape
```

```python
y.shape
```

```python
!pip install --upgrade imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_res,y_res = SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_res,y_res,test_size=0.20, random_state=42)
```

### Logistic Regression

```python
log = LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_pred1 = log.predict(X_test)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
precision_score(y_test,y_pred1)
```
